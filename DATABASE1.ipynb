```python
import sqlite3
```

```python
con=sqlite3.connect('students1.sqlite3')
```

```python
cur=con.cursor()
```

```python
create_table_query="""
CREATE TABLE studentt(
id int primary key,
name varchar not null,
address varchar not null,
college varchar not null,
age int not null,
gender varchar(1) not null,
faculty varchar not null
);

"""
cur.execute(create_table_query)
```

```python
students=[(12,'puja', 'lalitpur','ncit', 23, 'f', 'IT','TU'),
(13,'mamata', 'lalitpur','cosmos', 23, 'm', 'IT','TU'),
(14,'kabi', 'lalitpur','everest', 23, 'm', 'IT','TU'),
         ]

INSERT_RECORDS="""
    INSERT into studentt
    ('id','name','address','college','age','gender','faculty','university')
    values
    (?, ?, ? , ?, ?, ?, ?, ?);
    """
for i in  students:
    cur.execute(INSERT_RECORDS ,i) # single write at a time
    con.commit()
```

```python
# used executemany when there are multiple data
students=[(15,'maya', 'ktm','ncit', 23, 'f', 'IT','FWU'),
(16,'kamal', 'pkh','ashirbaad', 23, 'm', 'IT','TU'),
(17,'manju', 'mnr','araniko', 23, 'f', 'CE','PU'),
         ]

INSERT_RECORDS="""
    INSERT into studentt
    ('id','name','address','college','age','gender','faculty','university')
    values
    (?, ?, ? , ?, ?, ?, ?, ?);
    """

cur.executemany(INSERT_RECORDS ,students) # single write at a time
con.commit()
    
```
